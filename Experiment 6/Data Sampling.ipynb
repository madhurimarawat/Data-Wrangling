## Experiment 6

### Data Sampling: Randomly sample rows or columns from a dataset to create a smaller subset for analysis. 

### Loading Dataset

```python
# Import pandas library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset (for demonstration)
iris = sns.load_dataset('iris')

# Display the first few rows of the original dataset
print("Original Dataset (Iris):\n")
print(iris.head())
```

### 1. Row Sampling

```python
# Data Sampling: Randomly sample rows from the dataset
# Sampling 20% of the rows randomly without replacement
sampled_data_rows = iris.sample(frac=0.2, random_state=42)

# Display the sampled rows
print("\nSampled Rows (20% of total rows):\n")
print(sampled_data_rows.head())
```

### Plotting before and after sampling

```python
# Create a subplot with 1 row and 2 columns (side by side)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot original data (left side)
sns.scatterplot(ax=axes[0], data=iris, x='sepal_length', y='sepal_width', hue='species', palette='pastel')
axes[0].set_title("Original Iris Dataset")
axes[0].set_facecolor('white')  # Set teal background color
axes[0].grid(True)

# Plot sampled data (right side)
sns.scatterplot(ax=axes[1], data=sampled_data_rows, x='sepal_length', y='sepal_width', hue='species', palette='pastel')
axes[1].set_title("Sampled Iris Dataset (20%)")
axes[1].set_facecolor('white')  # Set teal background color
axes[1].grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

### 2. Column Sampling

```python
# Data Sampling: Randomly sample columns from the dataset
# Sampling 2 columns randomly from the dataset
sampled_data_columns = iris.sample(axis=1, n=2, random_state=42)

# Display the sampled columns
print("\nSampled Columns (2 random columns):\n")
print(sampled_data_columns.head())
```

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Data Sampling: Sample 2 columns randomly from the dataset
sampled_data_columns = iris.sample(axis=1, n=2, random_state=42)

# Create a subplot with 1 row and 2 columns (side by side)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot original data (left side) - all columns
sns.heatmap(iris.corr(numeric_only=True), ax=axes[0], annot=True, cmap='coolwarm')  # Added numeric_only=True
axes[0].set_title("Correlation Heatmap of Original Iris Dataset")
axes[0].set_facecolor('teal')  # Set teal background color

# Plot sampled data (right side) - only 2 randomly sampled columns
sns.heatmap(sampled_data_columns.corr(numeric_only=True), ax=axes[1], annot=True, cmap='coolwarm')  # Added numeric_only=True
axes[1].set_title("Correlation Heatmap of Sampled Iris Dataset (2 Columns)")
axes[1].set_facecolor('teal')  # Set teal background color

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
